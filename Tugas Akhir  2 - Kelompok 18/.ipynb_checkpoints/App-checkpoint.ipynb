```python
import os
import numpy as np
import pandas as pd
import flask
import pickle
import openpyxl
# KMeans Clustering
from sklearn.cluster import KMeans
# Preprocessing
from sklearn import preprocessing
from openpyxl import Workbook
from openpyxl import load_workbook
from flask import Flask, redirect, url_for, request, render_template

# creating instance of the class
app = Flask(__name__, template_folder='templates')

# to tell flask what url should trigger the function index()
@app.route('/')
@app.route('/index')
def index():
    return flask.render_template('index.html')
    
# prediction function
def ValuePredictor(to_predict_list):
    to_predict_list = list(map(float, to_predict_list))
    to_predict = np.array(to_predict_list).reshape(1,12)
    
    # Reading Dataset
    data_inventori = pd.read_excel("Dataset/Dataset18_Inventori.xlsx")
    data_inventoris = data_inventori[['Sales', 'Stock Out', 'PKM Exist', 'Avg Sales / Hari', 'MPKM Toko', 'PKM Toko', 'N+', 'Min Disp', 'Minor', 'Stock', 'DSI PKM', 'DSI OH']]
        
    # This is process to change the data to be array
    data_inventorinumpy = data_inventoris.to_numpy()
        
    # Define 2 variable for describing the maximum and the minimum of data
    mn = np.min(data_inventorinumpy, axis=0)  
    mx = np.max(data_inventorinumpy, axis=0) 
    
    data_predict = (to_predict - mn) / (mx - mn)
        
    loaded_model = pickle.load(open("model.pkl","rb")) # load the model
    result = loaded_model.predict(data_predict) # predict the values using loded model
    return result[0]

@app.route('/result', methods = ['POST'])
def result():
    if request.method == 'POST':
        to_predict_list = request.form.values()
        to_predict_list = list(map(str, to_predict_list))
        Plu = int(to_predict_list[0])
        Description = to_predict_list[1]
        TAG = to_predict_list[2]
        Sales = float(to_predict_list[3])
        Bulan = int(to_predict_list[4])
        Stock_Out = float(to_predict_list[5])
        Stock = float(to_predict_list[6])
        Avg_Sales_Hari = float(to_predict_list[7])
        Min_Disp = float(to_predict_list[8])
        Minor = float(to_predict_list[9])
        MPKM_Toko = float(to_predict_list[10])
        PKM_Toko = float(to_predict_list[11])
        NPlus = float(to_predict_list[12])
        DSI_PKM = float(to_predict_list[13])
        DSI_OH = float(to_predict_list[14])
        PKM_Exist = float(to_predict_list[15])
        
#----------------------------------------
        predict_list = [Sales, Stock_Out, Stock, Avg_Sales_Hari, Min_Disp, Minor, MPKM_Toko, PKM_Toko, NPlus, DSI_PKM, DSI_OH, PKM_Exist]       
        result = ValuePredictor(predict_list)
        
#----------------------------------------
        data_inventori = pd.read_excel("Dataset/Dataset18_Inventori.xlsx")
        data_inventoris = pd.DataFrame(data_inventori, columns=['Plu', 'Description', 'TAG', 'Bulan', 'Min Disp', 'Minor', 'Sales', 'Avg Sales / Hari', 'MPKM Toko', 'PKM Toko', 'N+', 'PKM Exist', 'Stock', 'Stock Out', 'DSI PKM', 'DSI OH'])
        # new dataframe with same columns
        df = pd.DataFrame([(Plu, Description, TAG, Bulan, Min_Disp, Minor, Sales, Avg_Sales_Hari, MPKM_Toko, PKM_Toko, NPlus, PKM_Exist, Stock, Stock_Out, DSI_PKM, DSI_OH)], columns=['Plu', 'Description', 'TAG', 'Bulan', 'Min Disp', 'Minor', 'Sales', 'Avg Sales / Hari', 'MPKM Toko', 'PKM Toko', 'N+', 'PKM Exist', 'Stock', 'Stock Out', 'DSI PKM', 'DSI OH'])
                           
        dataset_final = data_inventoris.append(df)
        dataset_final.set_index('Plu')
        writer = pd.ExcelWriter('Dataset/Dataset18_Inventori.xlsx')
        dataset_final.to_excel(writer, index=False)

        writer.save()
        
#----------------------------------------
        # Reading Dataset
        data_inventori = pd.read_excel("Dataset/Dataset18_Inventori.xlsx")
                
        # This is the process for retrieving positive value data
        data_inventori1 = data_inventori.loc[data_inventori['Stock']>=0]
        data_inventori1.reset_index(drop=True, inplace=True)
        
        # This is process of attributes allocation
        data_inventori2 = data_inventori1[['Sales', 'Stock Out', 'PKM Exist', 'Avg Sales / Hari', 'MPKM Toko', 'PKM Toko', 'N+', 'Min Disp', 'Minor', 'Stock', 'DSI PKM', 'DSI OH']]
        
        # This is process of scaling data 
        data_inventoriscaled = preprocessing.MinMaxScaler().fit_transform(data_inventori2)
        
        # This is process to change the Data to be Data Frame
        data_inventori3 = pd.DataFrame(data_inventoriscaled, columns=data_inventori2.columns[:13])
        
        # This is Process to select the attributes 
        X = data_inventori3[['Sales', 'Stock Out', 'PKM Exist', 'Avg Sales / Hari', 'MPKM Toko', 'PKM Toko', 'N+', 'Min Disp', 'Minor', 'Stock', 'DSI PKM', 'DSI OH']]

        # This is process to change the data to be array
        data_inventori2numpy = data_inventori2.to_numpy()
        
        # Define 2 variable for describing the maximum and the minimum of data
        mn = np.min(data_inventori2numpy, axis=0)  
        mx = np.max(data_inventori2numpy, axis=0)  
        
        # This is process to change the data to array
        data_inventori3numpy = data_inventori3.to_numpy()
        
        # This is process to return data to be as before
        data_inventori3 = (data_inventori3numpy * (mx - mn)) + mn

        # This is process to change the data tobe a dataframe
        data_inventori3 = pd.DataFrame(data_inventori3, columns=data_inventori2.columns[:12])

        # This is process to build model for k equals to 2
        model = KMeans(n_clusters=2, init = 'k-means++', random_state=None)
        # fit the model
        model.fit(X)
        
        # This is process to predict the values
        y_predicted = model.fit_predict(X)
    
        # This is process to add the new column to the dataframe
        data_inventori3['Cluster'] = y_predicted

        # serializing our model to a file called model.pkl
        import pickle
        pickle.dump(model, open("model.pkl","wb"))


#----------------------------------------        

        if float(result) == 0:
            prediction = 'Barang Termasuk kedalam Kategori "Laris"'
        elif float(result) == 1:
            prediction = 'Barang Termasuk kedalam Kategori "Kurang Laris"'
            
        return render_template("result.html",prediction=prediction, 
                               Plu = Plu,
                               Description = Description,
                               TAG = TAG,
                               Bulan = Bulan,
                               Sales = Sales, 
                               Stock_Out = Stock_Out, 
                               Stock = Stock, 
                               Avg_Sales_Hari = Avg_Sales_Hari, 
                               Min_Disp = Min_Disp, 
                               Minor = Minor, 
                               MPKM_Toko = MPKM_Toko, 
                               PKM_Toko = PKM_Toko, 
                               NPlus = NPlus, 
                               DSI_PKM = DSI_PKM,
                               DSI_OH = DSI_OH,
                               PKM_Exist = PKM_Exist) 
      
if __name__ == "__main__":
    app.run(debug=False) # use debug = False for jupyter notebook
```
